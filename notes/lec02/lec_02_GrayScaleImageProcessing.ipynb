# Gray Scale Image Processing

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
Image = np.array([[1,3,3,3],
                  [2,3,2,1],
                  [2,3,2,1],
                  [1,3,3,3],])
Image.shape
```

```python
def plot_image(img, L = 8, txt = 'Before HE' ):
    # K=256
    # img = (img *K )/ L
    # print(img)
    # plt.imshow(img, cmap='gray')
    plt.imshow(img, cmap='gray', vmin=0, vmax=L-1, interpolation='nearest')
    plt.title(txt)
    plt.axis('off')
    plt.savefig(txt.replace(' ','_') + '.png')
    plt.show()
    
plot_image(Image)
```

```python
Img2 = np.array([[2,7,7,7],
                  [4,7,4,2],
                  [4,7,4,2],
                  [2,7,7,7],])
plot_image(Img2,txt = 'After HE')
```
